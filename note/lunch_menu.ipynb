```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
pd.melt(df, id_vars=['A'], value_vars=['B'])
```

```python
mdf = df.drop(columns=['gmail', 'github', 'domain', 'vercel', 'role'])
mdf
```

```python
df_melt = mdf.melt(id_vars=['ename'], var_name='dt', value_name='menu_name')
df_melt
```

```python
# group by count ename 해보기 이때 - 등 결측치는 빼고 해보기
```

```python
sorted(df_melt['menu_name'].unique())
```

```python
#df_melt[df_melt['menu_name'].isin(['-', 'x', '<결석>'])]
df_melt[~df_melt['menu_name'].isin(['-', 'x', '<결석>'])]
```

```python
df_melt['menu_name'].isin(['-', 'x', '<결석>']).value_counts()
```

```python
not_na_df = df_melt[~df_melt['menu_name'].isin(['-', 'x', '<결석>'])]
```

```python
len(not_na_df)
```

```python
gdf=not_na_df.groupby('ename')['menu_name'].count().reset_index()
gdf
```

```python
not_na_df.groupby('ename')['menu_name'].count().reset_index().rename(columns={'menu_name': 'lunch_count'})
```

```python
#pdm add matplotlib
gdf.plot(x='ename', y='menu_name', kind='bar')
```
